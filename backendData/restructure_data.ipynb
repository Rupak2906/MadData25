```python
import math
def remove_nan_in_place(data):
    keys_to_remove = [k for k, v in data.items() if isinstance(v, float) and math.isnan(v)]
    for k in keys_to_remove:
        del data[k]

remove_nan_in_place(json_SG)

json_SG
```
